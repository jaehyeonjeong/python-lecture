```python
set01 = {1,2,3}
print (set01)
set02 = set("Hello")
print (set02)
set03 = set([1,2,3,4,5])
print (set03) # 리스트로부터 집합 생성

set04 = set((1,2,3,4,5))
set05 = set({4,5,6,7,8})
#교집합
insection_set = set04 & set05
print (insection_set)
union_set = set04 | set05
print (union_set)
minus_set = set04 - set05
print (minus_set)

set05.update([9,10])
print (set05)
set05.remove(10)
print (set05)
# set05.remove(100)  # 없는 값을 제거하려고 하면 오류 발생
set05.discard(100)  # 없는 값을 제거하려고 해도 오류 발생 안함
set05.clear()
print (set05)
```

```python
print(1==1)
print(1 == 2)
print(2 > 1)
print(2 < 1)
print(True and False)
print(True or False)
print(not True)

#truthy / falsy
print(bool("python"))
print(bool(""))  #빈 문자열은 False
print(bool(" "))  #공백 문자열도 True
print(bool([1,2,3]))  #빈 리스트가 아니므로 True
print(bool([]))  #빈 리스트는 False
print(bool((1,2,3)))  #빈 튜플이 아니므로 True
print(bool(()))  #빈 튜플은 False
print(bool(None))  #빈 튜플은 False
```

```python
print(bool({'name' : "정재헌"}))  #빈 딕셔너리가 아니므로 True
print(bool({}))  #빈 딕셔너리는 False
print(bool({None}))  #집합에 None이 있으므로 True
```

```python
# 3번 문제
nums = [1,2,3,4,5]
total = 0
for num in nums:
    total += num
print(total)

# 또는
sum_nums = sum(nums) # 내장 함수 sum() 사용
print(sum_nums) # 총합 출력
```

```python
# 4번 문제
nums = [3, 7, 2, 9, 4]
max_num = max(nums)  # 내장 함수 max() 사용
print(max_num)  # 최댓값 출력

# 또는
maxNum = nums[0]  # 리스트의 첫 번째 요소로 초기화
for num in nums:
    if num > maxNum:
        maxNum = num
print(maxNum)  # 최댓값 출력
```

```python
# 5번 문제
min_num = min(nums)  # 내장 함수 min() 사용
print(min_num)  # 최솟값 출력

# 또는
minNum = nums[0]  # 리스트의 첫 번째 요소로 초기화
for num in nums:
    if num < minNum:
        minNum = num
print(minNum)  # 최솟값 출력
```
