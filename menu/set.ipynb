```python
set01 = {1,2,3}
print (set01)
set02 = set("Hello")
print (set02)
set03 = set([1,2,3,4,5])
print (set03) # 리스트로부터 집합 생성

set04 = set((1,2,3,4,5))
set05 = set({4,5,6,7,8})
#교집합
insection_set = set04 & set05
print (insection_set)
union_set = set04 | set05
print (union_set)
minus_set = set04 - set05
print (minus_set)

set05.update([9,10])
print (set05)
set05.remove(10)
print (set05)
# set05.remove(100)  # 없는 값을 제거하려고 하면 오류 발생
set05.discard(100)  # 없는 값을 제거하려고 해도 오류 발생 안함
set05.clear()
print (set05)
```
