````python
odd = [1, 3, 5, 7, 9]
even = [2, 4, 6, 8, 10]
animals = ["cat", "dog", "monkey"]
blank = []
print(odd)
print(odd[0])
print(odd[0:2]) # a <= 출력되는 범위 < b

print(odd + even) # 리스트 연결(리스트 합치기)
print(odd * 2) # 리스트 반복
# tuple -> 불변객체, list -> 가변객체
odd[2] = 10
print(odd)
del odd[2]
print(odd)
print(odd[2])
odd.append(11)
print(odd)
for item in animals:
    print(item)
print(len(animals))
animals.append("lion")
print(animals)
animals.insert(1, "tiger")
print(animals)
popAnimal = animals.pop()   # 맨 마지막 요소를 꺼내서 반환
print(popAnimal)
print(animals.count("dog"))  # dog가 몇 개 있는지 세기
# animals.clear()               # 리스트 비우기
# print(animals)             # []
animals.remove("cat")  # cat 요소 삭제
print(animals)
````

````python
nums = [1,3,4,9,2,7,10,5]
nums.sort()  # 오름차순 정렬
print(nums)   #JS랑은 다르게 10이 맨 뒤로 감
nums.sort(reverse=True)  # 내림차순 정렬
print(nums)
# odd.extend(even)  # 리스트 확장
nums02 = odd+even
print(nums02)
print(odd)
# odd = odd + even
# odd += even
odd.extend(even)
print(odd)
````

````python
tuple01 = (1, 2, 3)   # 불변 객체 (추가, 삭제, 수정 불가) 오직 읽기 전용
print(tuple01)
tuple02 = (4,)   # int형 변수 tuple이 아님 -> 괄호로 감싼 것일 뿐
print(tuple02)
faketuple01 = (4)  # 요소가 1개인 튜플은 콤마(,)를 반드시 붙여야 함
print(type(tuple02))
print(type(faketuple01))
print(tuple01[0])
a, b, c = tuple01  # 언패킹)
print(a)
print(b)
print(c)
# 불변 객체를 쓰고 싶으면 tuple을, 가변 객체를 쓰고 싶으면 list를 사용하면 됨
````

````python
#list === list
#tuple === list (읽기 전용)
#dictionary === map (key-value 쌍으로 이루어진 자료구조) 수정(set) 가능
dic = {'name':'홍길동', 'age':30, 'address':'서울시', 1 :"001", 'name' : '김철수'} # key 중복 불가, 마지막에 선언된 값으로 저장
print (dic['name'])
print (dic)
print (dic[1])    # 정수로도 키를 쓸 수 있음
print (dic.keys())   # 키들만 출력
print (dic.values()) # 값들만 출력

for key in dic.keys(): # java랑은 다르게 key값을 간단히 출력
    print(key)  # key 출력 하나씩
````

````python
dic['email'] = 'jjh0306@gmail.com'
print(dic)
print(dic.get('emial'))  # 없는 키를 조회하면 None 반환
print(dic.get('email'))  # 없는 키를 조회하면 None 반환
print(dic['email'])  # 있는 키를 조회하면 값 반환
# print(dic['emill'])  # 없는 키를 조회하면 오류 발생
# print('email === ', dic.get('mail', 'not found'))  # 없는 키 조회 시 기본값 설정
mail = dic.get('mail', 'key not found')

print('mail === ', mail)
email = dic.pop('email')
print('email === ', email)
print(dic)
````
